```python
import json
import random
from collections import defaultdict
```

```python
input_file = 'data/TISER_test.json'

def sample_json(input_path, sample_size=500):
    dataset_groups = defaultdict(list)
    with open(input_path, 'r', encoding='utf-8') as f:
        for _, line in enumerate(f, 1):
            line = line.strip()
            if not line:
                continue
            try:
                item = json.loads(line)
                name = item.get('dataset_name')
                if name:
                    dataset_groups[name].append(item)
            except json.JSONDecodeError as e:
                print(e)

    output_items = []
    for name, all_items in dataset_groups.items():
        actual_sample_size = min(len(all_items), sample_size)

        random.seed(114)
        sampled_items = random.sample(all_items, actual_sample_size)
        output_items += sampled_items

    output_filename = f"data/TISER_test_random_3000.json"

    with open(output_filename, 'w', encoding='utf-8') as f:
        for entry in output_items:
            json.dump(entry, f, ensure_ascii=False)
            f.write('\n')

    print(f"Sample: {len(output_items)} -> Saved to: {output_filename}")
```

```python
sample_json(input_file)
```
